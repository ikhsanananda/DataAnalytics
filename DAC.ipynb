```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

test_df = pd.read_csv("Test.csv")
adult = []
ruang = []

for i in test_df['adults']:
  adult.append(i)

for i in test_df['room']:
  ruang.append(i)

arr_adult = np.array(adult).reshape((-1,1))
arr_ruang = np.array(ruang)

model = LinearRegression()
model.fit(arr_adult, arr_ruang)
model = LinearRegression().fit(arr_adult, arr_ruang)

r_square = model.score(arr_adult, arr_ruang)
print("Koefisien determinasi : ", r_square)
print("Intersep : ", model.intercept_)
print("Slope : ", model.coef_)
#room_predicted = model.intercept_ + model.coef_*arr_adult
#print("Perkiraan okupansi kamar : ", model.predict(arr_adult), sep="\n")

hasil_prediksi = []
for i in model.predict(arr_adult):
  hasil_prediksi.append(i)

perbandingan = (sum(hasil_prediksi)) 
print("Jumlah okupansi kamar di awal terhadap jumlah adult : ", sum(ruang))
print("Prediksi penambahan okupansi kamar terhadap jumlah adult : ", perbandingan)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
from pandas import DataFrame

dttrain = pd.read_csv('Test.csv')

city = []
for i in dttrain['buyer_city'] :  
  city.append(i)

dist = []
for i in dttrain['distance'] : 
  dist.append(i)


arr_city = np.array(city)
arr_dist = np.array(dist)

#plot without clustering
plt.style.use('seaborn')
plt.title("Scatter Plot Result Between Buyer_City and Distance on Test.csv", fontsize=20)
plt.xlabel('Buyer_City', fontsize=20)
plt.ylabel('Distance', fontsize=20)
plt.plot()
plt.scatter(arr_city, arr_dist)
plt.show()

#clustering K-Means
arr_dist = np.nan_to_num(arr_dist)

plt.style.use('seaborn')
plt.plot()
X = np.array(list(zip(arr_city, arr_dist))).reshape(len(arr_city),2)
colors=['blue', 'green', 'yellow', 'red', 'gray', 'orange']
markers=['o','v','s','p','h','^']

#number of cluster
K = 6
kmeans_model = KMeans(n_clusters=K).fit(X)

#plot with clustering
plt.plot()
#plt.title("Scatter Plot and Clustering K-Means Result Between Buyer_City and Distance on Test.csv", fontsize=20)
plt.xlabel('Buyer_City', fontsize=20)
plt.ylabel('Distance', fontsize=20)
for i, l in enumerate(kmeans_model.labels_):
  plt.plot(arr_city[i], arr_dist[i], color=colors[l], marker=markers[l],
           ls='None')
plt.show()
```
