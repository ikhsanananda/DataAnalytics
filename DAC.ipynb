```python
konek = 0
tidak = 0
hasil = []
for i in test_df['mobile']:
  if i==1 : 
    konek+=1
  elif i==0 : 
    tidak += 1

hasil.append(konek)
hasil.append(tidak)

labels = 'Terhubung ke internet', 'Tidak Terhubung ke internet'
colors = 'blue', 'green'

plt.pie(hasil, labels=labels, colors=colors, explode=(0, 0.2), autopct='%1.2f%%')
plt.axis('equal')
plt.title('Banyak device yang terhubung ke internet')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

test_df = pd.read_csv("Test.csv")
adult = []
ruang = []

for i in test_df['adults']:
  adult.append(i)

for i in test_df['room']:
  ruang.append(i)

arr_adult = np.array(adult).reshape((-1,1))
arr_ruang = np.array(ruang)

model = LinearRegression()
model.fit(arr_adult, arr_ruang)
model = LinearRegression().fit(arr_adult, arr_ruang)

r_square = model.score(arr_adult, arr_ruang)
print("Koefisien determinasi : ", r_square)
print("Intersep : ", model.intercept_)
print("Slope : ", model.coef_)
#room_predicted = model.intercept_ + model.coef_*arr_adult
#print("Perkiraan okupansi kamar : ", model.predict(arr_adult), sep="\n")

hasil_prediksi = []
for i in model.predict(arr_adult):
  hasil_prediksi.append(i)

perbandingan = (sum(hasil_prediksi)) 
print("Jumlah okupansi kamar di awal terhadap jumlah adult : ", sum(ruang))
print("Prediksi penambahan okupansi kamar terhadap jumlah adult : ", perbandingan)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import statistics as st

arr_kota = np.array(kota)
jarak=[]

#isi yang null dengan mean dari kolom
test_df['distance'] = test_df['distance'].replace('NaN', np.nan).fillna(st.mean(test_df['distance']))
for i in test_df['distance']:
  jarak.append(i)

arr_jarak = np.array(jarak)

#plot awal without clustering
plt.style.use('seaborn')
plt.title("Hasil Scatter Plot antara Buyer_City dan Distance pada file Test.csv", fontsize=20)
plt.xlabel('Buyer_City', fontsize=20)
plt.ylabel('Distance', fontsize=20)
plt.plot()
plt.xlim([0,60000])
plt.ylim([0,13000])
plt.scatter(arr_kota, arr_jarak)
plt.show()

plt.style.use('seaborn')
plt.plot()
X = np.array(list(zip(arr_kota, arr_jarak))).reshape(len(arr_kota),2)
colors=['blue', 'green', 'yellow', 'red', 'gray', 'orange']
markers=['o','v','s','p','h','^']

#number of cluster
K = 6
kmeans_model = KMeans(n_clusters=K).fit(X)

#plot akhir with clustering
plt.plot()
plt.title("Hasil Scatter Plot dan Clustering K-Means antara Buyer_City dan Distance pada file Test.csv", fontsize=20)
plt.xlabel('Buyer_City', fontsize=20)
plt.ylabel('Distance', fontsize=20)
for i, l in enumerate(kmeans_model.labels_):
  plt.plot(arr_kota[i], arr_jarak[i], color=colors[l], marker=markers[l],
           ls='None')
  plt.xlim([0,60000])
  plt.ylim([0,13000])
plt.show()
```
