```python
konek = 0
tidak = 0
hasil = []
for i in test_df['mobile']:
  if i==1 : 
    konek+=1
  elif i==0 : 
    tidak += 1

hasil.append(konek)
hasil.append(tidak)

labels = 'Terhubung ke internet', 'Tidak Terhubung ke internet'
colors = 'blue', 'green'

plt.pie(hasil, labels=labels, colors=colors, explode=(0, 0.2), autopct='%1.2f%%')
plt.axis('equal')
plt.title('Banyak device yang terhubung ke internet')
plt.show()
```

```python
part = 0
nopart = 0
hasil = []
for i in test_df['package']:
  if i==1 : 
    part +=1
  elif i==0 : 
    nopart += 1

hasil.append(part)
hasil.append(nopart)

labels = 'Merupakan part dari sebuah package', 'Tidak merupakan part dari sebuah package'
colors = 'blue', 'green'

plt.pie(hasil, labels=labels, colors=colors, explode=(0, 0.2), autopct='%1.2f%%')
plt.axis('equal')
plt.title('Banyak item yang merupakan part dari sebuah package')
plt.show()
```

```python
deal = 0
click = 0
hasil = []
for i in train_df['dealing']:
  if i==1 : 
    deal +=1
  elif i==0 : 
    click += 1

hasil.append(deal)
hasil.append(click)

labels = 'Membeli produk', 'Cuman melihat produk'
colors = 'blue', 'green'

plt.pie(hasil, labels=labels, colors=colors, explode=(0, 0.2), autopct='%1.2f%%')
plt.axis('equal')
plt.title('Banyak produk yang dibeli')
plt.show()
```
